## Step 1: Import Libraries

We only use a few simple libraries: `pandas` for tables, `numpy` for numbers/randomness,
and `matplotlib` (+ `matplotlib_venn`) for charts.

```python
# import the libraries we need
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib_venn import venn2

# this makes our random numbers repeatable every time we run the notebook
np.random.seed(42)
```

## Step 2: Generate the Dataset (200 Students)

The project asks us to generate a dataset with these columns:
`study_hours`, `attendance`, `group_discussion`, `previous_test_score`, `final_exam_pass`.

We create 200 student rows using simple random numbers, then save it as a CSV file.

```python
# how many students we want in our dataset
total_students = 200

# study_hours: random numbers around 10, between 1 and 20
study_hours = np.round(np.random.normal(10, 4, total_students), 1)
study_hours = np.clip(study_hours, 1, 20)

# attendance: random numbers around 75 percent, between 40 and 100
attendance = np.round(np.random.normal(75, 15, total_students), 1)
attendance = np.clip(attendance, 40, 100)

# previous_test_score: random numbers around 65, between 30 and 100
previous_test_score = np.round(np.random.normal(65, 15, total_students), 1)
previous_test_score = np.clip(previous_test_score, 30, 100)

print("Basic numeric columns generated for", total_students, "students")
```

```python
# Now we decide group_discussion (Yes/No) and final_exam_pass (Pass/Fail).
# We fix simple, clean counts so our probability calculations are easy to follow:
#   Yes + Pass = 90 students
#   Yes + Fail = 30 students
#   No  + Pass = 50 students
#   No  + Fail = 30 students
#   (Yes total = 120, No total = 80, Pass total = 140, Fail total = 60)

labels = ["Yes_Pass"] * 90 + ["Yes_Fail"] * 30 + ["No_Pass"] * 50 + ["No_Fail"] * 30
np.random.shuffle(labels)

group_discussion = []
final_exam_pass = []

for label in labels:
    if label == "Yes_Pass":
        group_discussion.append("Yes")
        final_exam_pass.append("Pass")
    elif label == "Yes_Fail":
        group_discussion.append("Yes")
        final_exam_pass.append("Fail")
    elif label == "No_Pass":
        group_discussion.append("No")
        final_exam_pass.append("Pass")
    else:
        group_discussion.append("No")
        final_exam_pass.append("Fail")

print("group_discussion and final_exam_pass columns are ready")
```

```python
# small realistic touch: students who passed tend to have a bit higher
# attendance and previous score than students who failed
for i in range(total_students):
    if final_exam_pass[i] == "Pass":
        attendance[i] = attendance[i] + 4
        previous_test_score[i] = previous_test_score[i] + 5
    else:
        attendance[i] = attendance[i] - 4
        previous_test_score[i] = previous_test_score[i] - 5

# keep attendance and score inside sensible ranges
attendance = np.clip(np.round(attendance, 1), 40, 100)
previous_test_score = np.clip(np.round(previous_test_score, 1), 30, 100)

# put everything into one table (DataFrame)
df = pd.DataFrame({
    "study_hours": study_hours,
    "attendance": attendance,
    "group_discussion": group_discussion,
    "previous_test_score": previous_test_score,
    "final_exam_pass": final_exam_pass
})

# save it as a csv file so it can be submitted along with this notebook
df.to_csv("expectation_decider_dataset.csv", index=False)

print("Dataset saved as expectation_decider_dataset.csv")
df.head(10)
```

## Section 1: Understanding the Basics

**What is Probability?**
Probability is a number between 0 and 1 that tells us how likely an event is to happen.
0 = impossible, 1 = certain. Formula: `P(Event) = favourable outcomes / total outcomes`.

**Key Terminology**
- **Experiment** – an action with an uncertain result (e.g. picking a student).
- **Sample Space** – all possible outcomes (all 200 students).
- **Event** – the outcome(s) we care about (e.g. "student passes").
- **Random Variable** – a number that comes out of a random event (e.g. number of passers).
- **Mutually Exclusive** – events that cannot happen together.
- **Independent Events** – one event does not affect the other's probability.
- **Conditional Probability** – probability of an event *given* another event already happened.

**Three example events from our dataset:**
1. A student studies more than 10 hours/week.
2. A student passes the final exam.
3. A student takes part in group discussion **and** passes the exam.

```python
# quick look at the dataset shape and column info
print("Rows, Columns:", df.shape)
df.info()
```

## Section 2: Empirical vs Theoretical Probability

- **Empirical probability** = based on real observed data (what actually happened).
- **Theoretical probability** = based on an assumption/logic, calculated without needing the real data.

We calculate one of each below.

```python
# EMPIRICAL PROBABILITY: probability of passing, based on real data
total_pass = (df["final_exam_pass"] == "Pass").sum()
total_students = len(df)

p_pass_empirical = total_pass / total_students
print("Students who passed:", total_pass, "out of", total_students)
print("Empirical P(Pass) =", round(p_pass_empirical, 3))
```

```python
# THEORETICAL PROBABILITY: assume attendance is equally likely anywhere from 40% to 100%
# theoretical P(attendance > 80) = favourable range / total range
favourable_range = 100 - 80
total_range = 100 - 40
p_attendance_theoretical = favourable_range / total_range

print("Theoretical P(attendance > 80%) =", round(p_attendance_theoretical, 3))

# compare with what the real data actually shows
actual_high_attendance = (df["attendance"] > 80).sum()
p_attendance_empirical = actual_high_attendance / total_students
print("Empirical P(attendance > 80%) from real data =", round(p_attendance_empirical, 3))
```

## Section 3: Random Variable & Probability Distribution

Let **X = number of students who pass**, out of 3 randomly picked students.
Since each student passes with probability `p = 0.70` (our empirical pass rate),
X follows a **Binomial Distribution**: X ~ Binomial(n=3, p=0.7).

Formula: `P(X=k) = C(3,k) * (p)^k * (1-p)^(3-k)`

```python
# simple function to calculate "n choose k" (combinations), written in plain steps
def n_choose_k(n, k):
    numerator = 1
    denominator = 1
    for i in range(k):
        numerator = numerator * (n - i)
        denominator = denominator * (i + 1)
    return numerator / denominator

n = 3
p = round(p_pass_empirical, 2)   # 0.70
q = 1 - p                        # probability of failing

# calculate probability for each value of X = 0, 1, 2, 3
distribution_table = []
for k in range(0, n + 1):
    prob = n_choose_k(n, k) * (p ** k) * (q ** (n - k))
    distribution_table.append([k, round(prob, 3)])

dist_df = pd.DataFrame(distribution_table, columns=["X (passes out of 3)", "P(X)"])
dist_df
```

```python
# check the probabilities add up to 1
print("Sum of all probabilities =", dist_df["P(X)"].sum())

# Mean and Variance of a Binomial distribution
mean_X = n * p
variance_X = n * p * q

print("Mean, E(X) =", mean_X)
print("Variance, Var(X) =", round(variance_X, 3))
```

## Section 4: Venn Diagram

- **Set A** = students who study more than 10 hours/week
- **Set B** = students who attend more than 80% of classes
- **A ∩ B** = students who satisfy both conditions

```python
# find how many students belong to each set
set_A = df[df["study_hours"] > 10]
set_B = df[df["attendance"] > 80]
set_A_and_B = df[(df["study_hours"] > 10) & (df["attendance"] > 80)]

count_A = len(set_A)
count_B = len(set_B)
count_A_and_B = len(set_A_and_B)

print("Students in A (study > 10 hrs):", count_A)
print("Students in B (attendance > 80%):", count_B)
print("Students in A and B (both):", count_A_and_B)
```

```python
# draw the venn diagram using the counts above
only_A = count_A - count_A_and_B
only_B = count_B - count_A_and_B

plt.figure(figsize=(6, 5))
venn2(subsets=(only_A, only_B, count_A_and_B),
      set_labels=("Study > 10 hrs/week (A)", "Attendance > 80% (B)"))
plt.title("Venn Diagram: Study Hours vs Attendance")
plt.show()
```

## Section 5: Contingency Table & Probability Calculations

We build a table comparing `group_discussion` (Yes/No) with `final_exam_pass` (Pass/Fail),
then calculate joint, marginal, and conditional probabilities from it.

```python
# contingency table: rows = group_discussion, columns = final_exam_pass
contingency_table = pd.crosstab(df["group_discussion"], df["final_exam_pass"], margins=True,
                                 margins_name="Total")
contingency_table
```

```python
# read the exact numbers we need from the table
yes_and_pass = contingency_table.loc["Yes", "Pass"]
yes_total = contingency_table.loc["Yes", "Total"]
pass_total = contingency_table.loc["Total", "Pass"]
grand_total = contingency_table.loc["Total", "Total"]

# Joint Probability: P(Yes AND Pass)
joint_probability = yes_and_pass / grand_total
print("Joint Probability P(Group Discussion=Yes AND Pass) =", round(joint_probability, 3))

# Marginal Probability: P(Pass)
marginal_probability = pass_total / grand_total
print("Marginal Probability P(Pass) =", round(marginal_probability, 3))

# Conditional Probability: P(Pass | Yes)
conditional_probability = yes_and_pass / yes_total
print("Conditional Probability P(Pass | Group Discussion=Yes) =", round(conditional_probability, 3))
```

## Section 6: Understanding Relationships

**Intuition:** `P(Pass | Yes)` means — out of only the students who did group discussion,
what fraction passed? We shrink the sample space from all 200 students down to just the 120
"Yes" students, then check the pass fraction inside that smaller group.

**Are the two events independent, dependent, or mutually exclusive?**
We compare `P(Pass)` (overall) with `P(Pass | Yes)` and `P(Pass | No)` below.

```python
# probability of pass given NO group discussion
no_and_pass = contingency_table.loc["No", "Pass"]
no_total = contingency_table.loc["No", "Total"]
p_pass_given_no = no_and_pass / no_total

print("P(Pass) overall            =", round(marginal_probability, 3))
print("P(Pass | Group Disc = Yes) =", round(conditional_probability, 3))
print("P(Pass | Group Disc = No)  =", round(p_pass_given_no, 3))

if round(conditional_probability, 3) == round(marginal_probability, 3):
    print("\nSince the values are EQUAL -> the events are INDEPENDENT")
else:
    print("\nSince the values are DIFFERENT -> the events are DEPENDENT")

if joint_probability == 0:
    print("Joint probability is 0 -> events are also MUTUALLY EXCLUSIVE")
else:
    print("Joint probability is NOT 0 -> events are NOT mutually exclusive")
```

## Section 7: Bayes' Theorem Application

Given (historical data, as stated in the project):
- `P(High Attendance | Pass) = 0.70`
- `P(High Attendance | Fail) = 0.40`
- `P(High Attendance) = 0.60`

We need to find `P(Pass | High Attendance)`.

**Step 1:** Find `P(Pass)` using the Law of Total Probability:
`P(HA) = P(HA|Pass) * P(Pass) + P(HA|Fail) * P(Fail)`

**Step 2:** Apply Bayes' Theorem:
`P(Pass | HA) = [P(HA|Pass) * P(Pass)] / P(HA)`

```python
# given values from the problem statement
p_ha_given_pass = 0.70
p_ha_given_fail = 0.40
p_ha = 0.60

# Step 1: solve for P(Pass) using the law of total probability
# p_ha = p_ha_given_pass * p_pass + p_ha_given_fail * (1 - p_pass)
# solving the equation for p_pass:
p_pass_prior = (p_ha - p_ha_given_fail) / (p_ha_given_pass - p_ha_given_fail)
p_fail_prior = 1 - p_pass_prior

print("Step 1 -> P(Pass) =", round(p_pass_prior, 3))
print("Step 1 -> P(Fail) =", round(p_fail_prior, 3))

# Step 2: apply Bayes theorem
p_pass_given_ha = (p_ha_given_pass * p_pass_prior) / p_ha
print("\nStep 2 -> P(Pass | High Attendance) =", round(p_pass_given_ha, 3))
print("This means a student with high attendance has about",
      str(round(p_pass_given_ha * 100, 1)) + "% chance of having passed.")
```
